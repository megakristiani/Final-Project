Beer is one of the oldest and most widely consumed  alcoholic drinks in the world, and the third most popular drink overall after water and tea. According to the World Bank 2015 report, Indonesia, on average drinking just 0.6 liters of alcohol (thejakartapost.com, 2018). Even though, Indonesia has produced commercial beer (Bintang, Stark Craft Beer, Anker Beer, Bali Hai) and and also own the traditional beer (Bir Pletok, Bir Kotjok, Bir, Jawa, etc). One of the problem for Indonesian beer makers is they they must have certificate issued by BPOM before selling their product in Indonesia. This final project will predict "is your beer allowed to be sold in Indonesia?"

- BeerID = Record ID
- Name = Name
- URL = Location of recipe webpage at https://www.brewersfriend.com
- Style = Type of brew
- StyleID = Numeric ID for type of brew
- Size(L) = Amount brewed for recipe listed
- OG = Specific gravity of wort before fermentation
- FG = Specific gravity of wort after fermentation
- ABV = Alcohol By Volume
- IBU = International Bittering Units
- Color = Standard Reference Method - light to dark ex. 40 = black
- BoilSize = Fluid at beginning of boil
- BoilTime = Time wort is boiled
- BoilGravity = Specific gravity of wort before the boil
- Efficiency = Beer mash extraction efficiency - extracting sugars from the grain during mash
- MashThickness = Amount of water per pound of grain
- SugarScale = Scale to determine the concentration of dissolved solids in wort
- BrewMethod = Various techniques for brewing
- PitchRate = Yeast added to the fermentor per gravity unit - M cells/ml/deg P
- PrimaryTemp = Temperature at the fermenting stage
- PrimingMethod = N
- PrimingAmount = Amount of priming sugar used
- UserId = UserId

1. IBU (International Bitterness Unit) preferences to the measurable facts about a beer to find that perfect sweet spot. Zero IBU (IBU 0) is a claim of sweetness present. (https://sixpoint.com/debunking-0-ibu-beer-facts-and-falsehoods-3-things-you-should-know-about-bitterness-in-your-beer/)

2. ABV (Alcohol By Volume) preferences to measurable amount of alcohol in volume. Zero ABV (ABV 0) is a claim of Non Alcoholic Beer (0 - 0.05% ABV). (https://www.drinkaware.co.uk/alcohol-facts/alcoholic-drinks-units/alcoholic-and-non-alcoholic-beers/)

3. Color preferences to measurable of color. Zero Color (Color 0) is a claim of Clear Near Beer (zero-color brew). (https://soranews24.com/2018/06/21/clear-near-beer-is-here-we-taste-test-japans-new-zero-alcohol-zero-color-brew%E3%80%90taste-test%E3%80%91/)

4. BoilTime preference to time of boiling wort.Zero minute boil time is a claim for beer whom the hops is added at flameoff or the kettle is off. (https://homebrew.stackexchange.com/questions/571/what-does-a-0-minute-boil-time-mean)

5. BoilGravity preference to preboil gravity. Zero Boil Gravity is a claim for beer whom the fermentable and unfermentable substance (wort and sugar) does not added before fermentation. (https://www.thespruceeats.com/original-gravity-of-beer-353235)

6. Efficiency preferences to measurement of potential fermentables converted into sugar in wort, this is a cumulative measurement of these losses taken right before the yeast is cast and fermentation begins. Zero efficiency is a claim for beer which have same volume as before fermentation. (http://howtobrew.com/book/section-2/what-is-malted-grain/mash-efficiency)

7. MashThickness preferences to measurement of amount of water per pound of grain or liquor-to-grist ratio. Zero mashthickness indicates that beer has liquor fewer or less than grain after fermentation. (https://byo.com/article/managing-mash-thickness/).

8. PitchRate preferencess to measurement of amount of yeast added to the fermentor per gravity unit (M cells/ml/deg P). Zero PitchRate indicates that beer has made with 0 days old yeast. (https://www.brewersfriend.com/yeast-pitch-rate-and-starter-calculator/)

9. PrimaryTemp preference to measurement of temperature at the fermenting stage (Celcius).

### Adding BPOM Column

I would like to add a feature named BPOM which contained number 1 - 6 based on beer type (Beer and Stout) and ABV (Alcohol By Volume) from Standar Keamanan Dan Mutu Minuman Beralkohol using index method because:
    - BPOM scoring the product accurately.
    - Clustering makes the qualified beer and stoud become unqualified, and unqualified beer and stoud become qualified.

BPOM Standard:
    - Beer (Color <24):
        ABV = 0-8%
    - Stout (Color >=24)
        ABV = 2-8%

```python
df['BPOM']=0
```

```python
beer=df[df['Color']<24]
idxb=beer.index
stout=df[df['Color']>=24]
idxs=stout.index
```

```python
a=beer[beer['ABV']<2]
idxa=a.index
b=beer[(beer['ABV']>=2) & (beer['ABV']<=8)]
idxb=b.index
c=beer[beer['ABV']>8]
idxc=c.index
```

```python
d=stout[stout['ABV']<2]
idxd=d.index
e=stout[(stout['ABV']>=2) & (stout['ABV']<=8)]
idxe=e.index
f=stout[stout['ABV']>8]
idxf=f.index
```

```python
df.loc[idxa,'BPOM']=1
df.loc[idxb,'BPOM']=2
df.loc[idxc,'BPOM']=3
df.loc[idxd,'BPOM']=4
df.loc[idxe,'BPOM']=5
df.loc[idxf,'BPOM']=6
```

# 1. BPOM Prediction

The first prediction is BPOM Prediction, this results is number of group between 1-6.

Result:
    - 1,2,5 : the beer can be sold in Indonesia
    - 3,4,6 : the beer can not be sold in Indonesia

```python
df['BPOM'].value_counts(normalize=True)
```

As we can see, the data are so imbalance. I need to apply SMOTE to make it balance.

### SMOTE

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='ABV', y='Color', hue='BPOM', data=df, alpha=0.3, palette='hot_r')
plt.title('Prediction')
plt.show()
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=101)

# setting target and feat
X=df.drop(['BPOM','Name','Style'], axis=1)
y=df['BPOM']
X=pd.get_dummies(data=X,
                columns=['SugarScale','BrewMethod'],
                drop_first=True)
```

```python
X_sm, y_sm =sm.fit_sample(X, y)
```

```python
# target
X_sm['BPOM']=y_sm
```

```python
X_sm['BPOM'].value_counts(normalize=True)
```

Now the data are balance. Next, i will apply SVM Method.

```python
# dfsm for data that has been apllied with SMOTE
dfsm=X_sm
dfsm['BPOM']=y_sm
```

```python
dfsm.head()
```

### Target and Features

```python
# Setting target and features
X=dfsm[['Color','ABV']]
y=dfsm['BPOM']
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30)
```

I would like to apply SVM, Decision Tree Classifier, Random Forest Classifier, and XGBOOST, and choose which one is the best method to predict the result.

### SVM

```python
from sklearn import svm
svc = svm.SVC(kernel='poly', degree=6, random_state=101).fit(X_train,y_train)
svc
```

```python
y_pred = svc.predict(X_test)
pred_train = svc.predict(X_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_train, pred_train))
print(classification_report(y_train, pred_train))
```

```python
from eli5 import show_weights
from eli5.sklearn import PermutationImportance

permutation = PermutationImportance(svc,scoring='accuracy', random_state=101).fit(X_test, y_test)
show_weights(permutation, feature_names=list(X_test.columns))
```

### Decission Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(random_state=101)
dtree.fit(X_train,y_train)
```

```python
# Predict Proba
dtree.predict_proba(X_test)
```

```python
print(confusion_matrix(y_test, dtree.predict(X_test)))
print(classification_report(y_test, dtree.predict(X_test)))
```

```python
print(confusion_matrix(y_train, dtree.predict(X_train)))
print(classification_report(y_train, dtree.predict(X_train)))
```
